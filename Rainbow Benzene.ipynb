```python
clear() # reset the canvas 
speed(10) # 13 is the fastest 1 is the slowest 
jumpto(0,0) # set the starting position to 0,0 think of it like a graphing axis 
colors = ['red', 'purple', 'blue', 'green', 'orange', 'yellow'] # different colors we will use for the drawing 
bgcolor('black') # set the background color to black 
for x in range(300):
    pencolor(colors[x%6]) # note that 0/6 = 0 with a remainder of 0. 1/6= 0 with a remainder of 1 also known as 1/6 = .166666 
    width(x//100 + 1) #performs floor divistion 
    forward(x) # go forward by x units at whatever angle we are currently at
    left(59) #turn counter clockwise 59 degrees
```
